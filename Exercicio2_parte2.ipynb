```python
#Abre o arquivo
def ler_arquivo_csv(arquivo = '', delimter = ','):
    content = '';
    try:
      with csv.reader(arquivo, 'r') as file:
        content = file.read()
        content.encode('UTF-8')
        print(content)
        file.close
    except FileNotFoundError:
        print("Arquivo não existe")
    except e:
        print(e.errno)
    return content  
```

```python
working_at_computer = []
standing_up = []
standing = []
walking = []
going_updown_stairs = []
walking_and_talking = []
talking_while_standing = []

idxes = []

with open(file, newline='') as csvfile:
     spamreader = csv.reader(csvfile, delimiter=',')
     for row in spamreader:
       idxes.append(int(float(row[0])) +1 )
       if (row[4] == '1'):
         working_at_computer.append(row)
       elif (row[4] == '2'):
         standing_up.append(row)
       elif (row[4]=='3'):
         standing.append(row)
       elif (row[4]=='4'):
         walking.append(row)
       elif (row[4]=='5'):
         going_updown_stairs.append(row)
       elif (row[4]=='6'):
         walking_and_talking.append(row)
       elif (row[4]=='7'):
         talking_while_standing.append(row)


```

```python


w1, w2, w3 = [],[],[]
for vetor1 in working_at_computer:
  w1.append(int(vetor1[1]))
  w2.append(int(vetor1[2]))
  w3.append(int(vetor1[3]))

#loan_amount = cur.fetchall()
#loan_amount = np.array(w1,dtype='int') #setting up the array in numpy
x ,y  = np.unique(w1, return_counts=True) # counting occurrence of each loan
plt.scatter(x,y)

x ,y = np.unique(w2,return_counts=True)
plt.scatter(x,y)

x ,y = np.unique(w3,return_counts=True)
plt.scatter(x,y)
plt.title("working_at_computer")
plt.show()


su1, su2, su3 = [],[],[]
for vetor1 in standing_up:
  su1.append(int(vetor1[1]))
  su2.append(int(vetor1[2]))
  su3.append(int(vetor1[3]))

x ,y  = np.unique(su1, return_counts=True) # counting occurrence of each loan
plt.scatter(x,y)

x ,y = np.unique(su2,return_counts=True)
plt.scatter(x,y)

x ,y = np.unique(su3,return_counts=True)
plt.scatter(x,y)
plt.title("standing_up")
plt.show()

s1, s2, s3 = [],[],[]
for vetor1 in standing:
  s1.append(int(vetor1[1]))
  s2.append(int(vetor1[2]))
  s3.append(int(vetor1[3]))

x ,y  = np.unique(s1, return_counts=True) # counting occurrence of each loan
plt.scatter(x,y)

x ,y = np.unique(s2,return_counts=True)
plt.scatter(x,y)

x ,y = np.unique(s3,return_counts=True)
plt.scatter(x,y)
plt.title("standing")
plt.show()


#####

wa1, wa2, wa3 = [],[],[]
for vetor1 in walking:
  wa1.append(int(vetor1[1]))
  wa2.append(int(vetor1[2]))
  wa3.append(int(vetor1[3]))

x ,y  = np.unique(wa1, return_counts=True) # counting occurrence of each loan
plt.scatter(x,y)

x ,y = np.unique(wa2,return_counts=True)
plt.scatter(x,y)

x ,y = np.unique(wa3,return_counts=True)
plt.scatter(x,y)
plt.title("walking")
plt.show()

#####

gus1, gus2, gus3 = [],[],[]
for vetor1 in going_updown_stairs:
  gus1.append(int(vetor1[1]))
  gus2.append(int(vetor1[2]))
  gus3.append(int(vetor1[3]))

x ,y  = np.unique(gus1, return_counts=True) # counting occurrence of each loan
plt.scatter(x,y)

x ,y = np.unique(gus2,return_counts=True)
plt.scatter(x,y)

x ,y = np.unique(gus3,return_counts=True)
plt.scatter(x,y)
plt.title("going_updown_stairs")
plt.show()

#####

wat1, wat2, wat3 = [],[],[]
for vetor1 in walking_and_talking:
  wat1.append(int(vetor1[1]))
  wat2.append(int(vetor1[2]))
  wat3.append(int(vetor1[3]))

x ,y  = np.unique(wat1, return_counts=True) # counting occurrence of each loan
plt.scatter(x,y)

x ,y = np.unique(wat2,return_counts=True)
plt.scatter(x,y)

x ,y = np.unique(wat3,return_counts=True)
plt.scatter(x,y)
plt.title("walking_and_talking")
plt.show()


#####

tws1, tws2, tws3 = [],[],[]
for vetor1 in talking_while_standing:
  tws1.append(int(vetor1[1]))
  tws2.append(int(vetor1[2]))
  tws3.append(int(vetor1[3]))

x ,y  = np.unique(tws1, return_counts=True) # counting occurrence of each loan
plt.scatter(x,y)

x ,y = np.unique(tws2,return_counts=True)
plt.scatter(x,y)

x ,y = np.unique(tws3,return_counts=True)
plt.scatter(x,y)
plt.title("talking_while_standing")
plt.show()

```
